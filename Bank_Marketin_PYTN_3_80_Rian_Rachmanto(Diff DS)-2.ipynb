**Data types from data set consists of 11 non-numeric data (catageory) and 10 numeric data. Before we employ all the data for machine learning, the non-numeric data types need to be converted first to numeric data types. We will use labelencoder for converting non numeric data to numeric data**


```python
from sklearn.preprocessing import LabelEncoder


df1=df.apply(LabelEncoder().fit_transform)
df1
```

```python
df1.tail(10)
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
sns.heatmap(data=df1.corr(), annot=True, cmap='viridis')
```

**Based on the heat Map above, we can see that emp.var.rate, euribor3m and nr.employed show high number of correlation. Therefore we can omit nr.employed, emp.var.rate and duration (based on the information from UCI, duration is not needed for ML model)**

```python
df2=df1.drop(['nr.employed','duration','emp.var.rate'],axis=1)
df2.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data=df2.corr(), annot=True, cmap='viridis')
```

```python
X =df2.drop(['y'],axis=1)
y = df2[['y']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_cv, y_train, y_cv = train_test_split(X, y, test_size = 0.2, random_state=0)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(solver='liblinear', random_state=0).fit(X_train,y_train)
```

```python
pred_cv=logreg.predict(X_cv)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print(accuracy_score(y_cv,pred_cv))
matrix=confusion_matrix(y_cv,pred_cv)
print(matrix)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
pred_cv2=knn.predict(X_cv)
```

```python
print(accuracy_score(y_cv,pred_cv2))
matrix2=confusion_matrix(y_cv,pred_cv2)
print(matrix2)
```

**SVM**

```python
from sklearn import svm
svm_model=svm.SVC()
svm_model.fit(X_train,y_train)
```

```python
svm.SVC(C=1.0, break_ties=False, cache_size=200, class_weight=None,
    coef0=0.0,decision_function_shape='ovr', degree=3,
    gamma='scale', kernel='rbf', max_iter=-1,
    probability=False, random_state=None, shrinking=True, 
    tol=0.001,verbose=False)
```

```python
pred_cv3=svm_model.predict(X_cv)
```

```python
print(accuracy_score(y_cv,pred_cv3))
matrix3=confusion_matrix(y_cv,pred_cv3)
print(matrix3)
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
pred_cv4=gnb.predict(X_cv)
```

```python
print(accuracy_score(y_cv,pred_cv4))
matrix4=confusion_matrix(y_cv,pred_cv4)
print(matrix4)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred_cv5=dt.predict(X_cv)
```

```python
print(accuracy_score(y_cv,pred_cv5))
matrix5=confusion_matrix(y_cv,pred_cv5)
print(matrix5)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rfs=RandomForestClassifier()
rfs.fit(X_train,y_train)
pred_cv6=rfs.predict(X_cv)
```

```python
print(accuracy_score(y_cv,pred_cv6))
matrix6=confusion_matrix(y_cv,pred_cv6)
print(matrix6)
```

**Comparing All Alogritm**

```python
print('Logistic Regression Accuracy: ', accuracy_score(y_cv,pred_cv))
print('K Nearest Neighbors: ', accuracy_score(y_cv,pred_cv2))
print('SVM: ', accuracy_score(y_cv,pred_cv3))
print('Naive Bayes: ', accuracy_score(y_cv,pred_cv4))
print('Decision Tree: ', accuracy_score(y_cv,pred_cv5))
print('Random Forest: ', accuracy_score(y_cv,pred_cv6))
```

Based on the comparison, Logistic Regression gives the best result
